```python
import matplotlib.pyplot as plt

verified = df['verified'] == True
not_verified= df['verified'] == False

fig, ax = plt.subplots()
ax.pie([sum(verified), sum(not_verified)], labels=['verified', 'Not verified'], autopct='%1.1f%%')
ax.axis('equal')

plt.show()
```

```python
likes = df['likes']
retweets = df['retweets']

plt.plot(likes, label='Likes')
plt.plot(retweets, label='Retweets')

plt.xlabel('Tweets')
plt.ylabel('Count')

plt.title('Likes and Retweets')

plt.legend()
plt.show()
```

```python
plt.style.use('seaborn')
likes = df['likes']
retweets = df['retweets']
followers = df['followers']
ratio = df['likes'] / df['retweets']
plt.scatter(followers, likes, c=ratio, cmap='summer',
edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Like Retweet Ratio')
plt.xscale('log')
plt.yscale('log')
plt.title('Analysis')
plt.xlabel('Followers')
plt.ylabel('Total Likes')
```
